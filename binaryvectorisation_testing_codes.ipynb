```python
import requests
from bs4 import BeautifulSoup
import gensim
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
def preprocess_text(text):
          # Remove HTML tags
          text = BeautifulSoup(text, "html.parser").get_text()
          # Convert to lowercase
          text = text.lower()
          # Remove non-alphanumeric characters
          text = re.sub(r'[^a-zA-Z0-9\s]', '', text)



          # Tokenize the text into words
          words = word_tokenize(text)



          # Remove stopwords
          words = [word for word in words if word not in stopwords.words('english')]



          # Lemmatize the words
          lemmatizer = WordNetLemmatizer()
          words = [lemmatizer.lemmatize(word) for word in words]



          return words





URLs = ["https://find-energy-certificate.service.gov.uk/energy-certificate/9243-1656-8157-5614-7032", "https://find-energy-certificate.service.gov.uk/energy-certificate/9442-4067-0631-2608-7495"]
      #url = "https://find-energy-certificate.service.gov.uk/energy-certificate/9243-1656-8157-5614-7032"




cleanedTexts = []
for url in URLs:
          response = requests.get(url)

          content = response.text
          soup = BeautifulSoup(content, 'html.parser')

          text = soup.get_text()

          # Preprocess the text
          cleaned_text = preprocess_text(text)

          cleanedTexts.append(cleaned_text)

```

```python
# Create a CountVectorizer object with binary=True
vectorizer = CountVectorizer(binary=True)

# Fit and transform the sentences to obtain binary vectors
binary_vectors = vectorizer.fit_transform(cleaned_text).toarray()

# Get the unique words and their corresponding indices
word_to_index = vectorizer.vocabulary_
index_to_word = {index: word for word, index in word_to_index.items()}

# Print the binary vectors and their corresponding words
for i, binary_vector in enumerate(binary_vectors):
    sentence =cleaned_text[i]
    words = sentence.split()
    print("Sentence:", sentence)
    print("Binary Vector:")
    for word, binary_value in zip(words, binary_vector):
        print(f"{word}: {binary_value}")

```
